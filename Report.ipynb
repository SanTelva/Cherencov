# Краткие и наглядные итоги моделирования и обработки данных
Здесь приведу много красивых графиков для электронов. Для протонов результаты обработки постараюсь сделать поскорее, тут всё зависит только от ресурсов компьютера и скорости интернета, сама программа уже написана и работает. 
## Световыход (самое интересное)
Ниже представлен график для распределения количества фотонов по результатам 10000 симуляций. 
![light_density1](Result/light_density1.png "Распределение зарегистрированых фотонов")

Здесь показано распределение именно __зарегистрированных__ фотонов. Не стал делать логарифмический масштаб. На мой взгляд, тут всё достаточно красноречиво и понятно без лишних слов. Чем больше энергия первичного электрона, тем больше фотонов появится в выходном окне. Зависимость среднего количества таких фотонов линейна и видна из следующего рисунка:

![light_dependence](Result/light_dependence.png "Зависимость от энергии зарегистрированых электронов")

Здесь показана зависимость среднего количества пойманных фотонов от энергии первичного электрона. Считаю, что это линейная зависимость. Доверительные интервалы для погрешности выбраны как $1.2\sigma$. Также абсолютно очевидно, что электроны с энергией меньше 1,5 МэВ прибор будет регистрировать *очень* плохо.

Взглянем на полную картину и найдём эффективность детектора: а со скольких фотонов, собственно, у нас есть световыход?

![cercount_dependence](Result/cercount_dependence.png "Зависимость от энергии количества черенковских фотонов")

В идеальных условиях количество родившихся черенковских частиц пропорционально энергии первичной частицы, однако в реальности эта линейная зависимость должна, как мне кажется, заваливаться в силу энергетических потерь частицы, а также неучтённых в нашей модели сцинтилляционных свойств. Вспоминается любимая фраза Свертилова, о том, что в природе нет ничего идеального.

В завершение можно поделить один ряд данных на другой и увидеть эффективность прибора:

![ratio](Result/ratio.png "")

Здесь нужно, серьёзно говоря, указать погрешности, но я поленился их правильно считать. Так или иначе, видим, что эффективность есть приблизительно постоянная величина, в районе 4%. Подъём в начале можно объяснить скорее большой погрешностью и дискретностью величины. К тому же, как было сказано выше, эти энергии нас почти не устраивают.

## Длина пути электрона и глубина проникновения

Сразу скажу, что глубина проникновения $-$ некорректное название, нужно что-то другое, потому что под глубиной обычно подразумевается координата $z$, а здесь расстояние, пройденное от точки вхождения до точки останова. Длина пути  $-$ это длина всего трека первичного электрона внутри кристалла 

![length_density](Result/length_density.png "Распределение длин пути электронов")

Всё понятно и без всяких объяснений. Для наглядности можно рядом положиь график плотности глубины.

![distance_density](Result/distance_density.png "Распределение длин глубин электронов")

Из разницы этих графиков можно как-нибудь судить об интенсивности флуктуаций электрона внутри кристалла, но я не знаю, как. В завершение представлю графики зависимости средних величин от энергии, которые могут быть также аппроксимированы линейной функцией или стихающим ростом:

![length_dependence](Result/length_dependence.png "Зависимость длин пути электронов")

## Энерговыделение

Сейчас две самые неоднозначные картинки для, соответственно, распределения энерговыхода и его зависимости от энергии первичной частицы. Как мы помним, на разных компьютерах программа давала разные результаты, у нас с вами вроде как всё идентично. 

![edep_density](Result/edep_density.png "Распределение энерговыхода")

![edep_dependence](Result/edep_dependence.png "Зависимость энерговыхода от энергии")

Во-первых, странно считается и нормируется плотность (максимум больше единицы). Функцию писал не самостоятельно, воспользовался доступным инструментарием. Попробую с этим разобраться, но ничего не обещаю.

Во-вторых, интересные зависимости для высокоэнергичных частиц. Если для "медленных" частиц ярко выражен пик в районе их первичной энергии (вся энергия отдаётся кристаллу), то для "быстрых", видимо, много энергии уходит на ионизацию (см. маленькие пики приблизительно на половине первичной энергии), при этом частица то ли не выходит из кристалла, то ли я её забываю поймать на выходе. Странно это. 

С зависимостью всё яснее, там идёт насыщение.
