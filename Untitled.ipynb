```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

class Electron():
    def __init__(self, xx, yy, zz, en, ID):
        self.x, self.y, self.z = xx, yy, zz
        self.energy = en
        self.id = ID
fin = open("particles-tracks", "r")
prim = Electron(0, 0, 0, 0, 1)
track = 0
depth = 0
penetration = False
part = ''
ID = 1
extruders = set()
cercount = 0
run = 1

result = []

x, y, z, en1, xx, yy, zz, en = 0, 0, 0, 0, 0, 0, 0, 0
for string in fin.readlines()[1:]:
    #print(string[:-1].split("\t"))
    if string == "\n" or "":
        depth = ((prim.x - x) ** 2 + (prim.y - y) ** 2 + (prim.z - z) ** 2) ** 0.5
        result.append((run, depth, track, len(extruders)))
        #print("Run", run)
        run += 1
        #print('Length of path is', track)
        track = 0

        #print("Depth is ", depth)
        #print(len(extruders), "Cerenkov photons passed through border\n")
        #print(extruders)
        extruders = set()
        penetration = False
    else:   
        try:
            part, x, y, z, en1, xx, yy, zz, en, ID = string[:-1].split("\t")
        except:
            pass
            #print(string)
        x, y, z, en1, xx, yy, zz, en = map(float, (x, y, z, en1, xx, yy, zz, en))
        ID = int(ID)
        if ID == 1:
            if penetration:
                track += ((xx - x) ** 2 + (yy - y) ** 2 + (zz - z) ** 2)**0.5
            elif zz == -20.0:
                prim.x, prim.y, prim.z = xx, yy, zz #beginning of penetration
                penetration = True
            prim.energy = en
            #print(x, y, z, en1, xx, yy, zz, en, track, sep ="\t")
            #if prim.energy == 0 and en1 == 0:   
        elif part == "opticalphoton" and zz == 40.0 and (xx * 2 + yy **2 <= 1) and ID not in extruders:
            #print("ID", ID, "photon passed through border")
            extruders.add(ID)
else:
    depth = ((prim.x - x) ** 2 + (prim.y - y) ** 2 + (prim.z - z) ** 2) ** 0.5
    result.append((run, depth, track, len(extruders)))
    #print("Run", run)
    run += 1
    #print('Length of path is', track)
    track = 0
    #print("Depth is ", depth)
    #print(len(extruders), "Cerenkov photons passed through border\n")
    #print(extruders)
    extruders = set()
    penetration = False
fin.close()
df = pd.DataFrame(result)
df.index = df[0]
edf = pd.read_csv("logoutput", sep = "\t")
edf.index = edf['Run']
df['Edep'] = edf['Edep']
df['CerCount'] = edf['CerCount']
df.columns = ['Run', 'Distance', 'Length', 'Counts', 'Edep', 'CerCount']
df.head()
```

```python
df.to_csv("15MeVData.csv")
```
